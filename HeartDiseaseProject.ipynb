```python
df["Diagnosis_binary"] = df["Diagnosis"].apply(lambda x: 1 if x > 0 else 0)
print(df["Diagnosis_binary"].value_counts())
```

```python
features = ["RestingBP", "Cholesterol", "FastingBS", "MaxHR"]
X = df[features]
y = df["Diagnosis_binary"]
```

```python
print(X.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve

y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Confusion Report:\n", classification_report(y_test, y_pred))

# ROC Curve
fpr, tpr, threshold = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
print("AUC Score:", roc_auc_score(y_test, y_pred_prob))
```

```python
feature_importance = pd.Series(model.coef_[0], index=features)
print(feature_importance.sort_values(ascending=False))
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

df["Clusters"] = cluster_labels
```

```python
df_clustered = df.copy()
df_clustered["Cluster"] = cluster_labels

sns.pairplot(df_clustered[["RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Cluster"]],
            hue="Cluster",
            plot_kws={'alpha':0.5},
            diag_kind='kde')
plt.suptitle("Pairwise Relationship Colored by K-means Cluster", y=1.05)
plt.show()
```

```python
# Cross_tabulation
pd.crosstab(df_clustered["Diagnosis_binary"], df_clustered["Cluster"],
          rownames=["Actual Heart Disease"], colnames=["Cluster Label"])
```

```python
df_clustered.groupby("Cluster")[["RestingBP", "Cholesterol", "FastingBS", "MaxHR"]].mean()
```

```python
X_pair = df[["RestingBP", "MaxHR",]].fillna(df[["RestingBP", "MaxHR"]].mean())
X_scaled_pair = scaler.fit_transform(X_pair)

for n_clusters in range(3, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled_pair)
    centroids = kmeans.cluster_centers_

    plt.figure(figsize=(7, 5))
    plt.scatter(X_scaled_pair[:, 0], X_scaled_pair[:, 1], c=cluster_labels, cmap='viridis', alpha=0.6, label="Patients")
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.title(f"K-means Clustering (n_clusters={n_clusters})")
    plt.xlabel("RestingBP (scaled)")
    plt.ylabel("MaxHR (scaled)")
    plt.legend()
    plt.show()
```
